# Determining the optimum number of clusters

```python
# Finding the optimum number of clusters
x = df.iloc[:,0:4].values
x
```

```python
from sklearn.cluster import KMeans
wcss = [] # WCSS is the sum of squares of distances of each 
          # data point in all cluster to their respective centroids
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',
                   max_iter=300,n_init=10,random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
wcss    
```

 We are going to use ELBOW method to find value of k

```python
# Find value of k
plt.plot(range(1,11),wcss, marker='o',markerfacecolor='red')
plt.title('Elbow method')
plt.xlabel(' no of clusters')
labels = ['Number of clusters', 'Wcss']
plt.ylabel('wcss') # within cluster sum of squares / wcss is low for higher no. of clusters
plt.legend(labels=labels)
plt.show()
```

We can clearly see why it is called 'the elbow method ' from the above graph, the optimum clusters
is where the elbow occurs. This is when the within clusters sum of squares (WCSS) doesn't decrease 
significantly with every iteration

```python
# From this we choose the number of clusters as 3
# Clustering
kmeans = KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=0)
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

# Plotting the Clusters

```python
# Visualising the clusters - On the first two columns
plt.scatter(x[identified_clusters == 0, 0], x[identified_clusters == 0 ,1],
            s=100,c='red',label='Iris-setosa')
plt.scatter(x[identified_clusters== 1,0], x[identified_clusters == 1,1],
           s=100,c='blue',label='Iris-versicolour')
plt.scatter(x[identified_clusters==2,0],x[identified_clusters==2,1],
           s=100,c='purple',label='Iris-virginica')
# Plotting the centroids of clusters
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            s=100,c='orange',label='Centroids')
plt.title('Distribution of points in Cluster')
plt.legend()
```

 3D scatterplot using matplotlib

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection='3d')
plt.scatter(x[identified_clusters == 0, 0], x[identified_clusters == 0 ,1],
            s=50,c='red',label='Iris-setosa')
plt.scatter(x[identified_clusters== 1,0], x[identified_clusters == 1,1],
           s=50,c='blue',label='Iris-versicolour')
plt.scatter(x[identified_clusters==2,0],x[identified_clusters==2,1],
           s=50,c='purple',label='Iris-virginica')
# Plotting the centroids of clusters
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            s=50,c='orange',label='Centroids')
plt.title('Distribution in 3D')
```

Additing the prediction to the dataset

```python
df_w_clusters = df.copy()
df_w_clusters['Cluster'] = identified_clusters
# Examine it
df_w_clusters.head()
```
